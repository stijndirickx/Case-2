```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


# Example data creation
df = pd.read_csv('Processed/data_clean.csv')

# Plot the distribution of unit_sales
plt.figure(figsize=(10, 6))
sns.histplot(df['unit_sales'], kde=True)
plt.title('Distribution of Unit Sales')
plt.xlabel('Unit Sales')
plt.ylabel('Frequency')
plt.show()

# Identify outliers using z-score
from scipy import stats
import numpy as np

# Calculate z-scores
z_scores = np.abs(stats.zscore(df['unit_sales']))

# Set a threshold for outliers
threshold = 3

# Find outlier indices
outlier_indices = np.where(z_scores > threshold)[0]

# Print outlier values and indices
if len(outlier_indices) > 0:
    #print("Outlier(s) found:")
    for idx in outlier_indices:
        print(f"Index: {idx}, Value: {df['unit_sales'][idx]}")
else:
    print("No outliers found.")

# Plot the data with outliers highlighted
plt.figure(figsize=(10, 6))
sns.histplot(df['unit_sales'], kde=True)
plt.scatter(outlier_indices, df['unit_sales'][outlier_indices], color='red', label='Outliers')
plt.title('Distribution of Unit Sales with Outliers')
plt.xlabel('Unit Sales')
plt.ylabel('Frequency')
plt.legend()
plt.show()
```
