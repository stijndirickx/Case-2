# Did you notice negative sales?  

Yes, we noticed negative sales. 138 times to be exact.

# How did you handle these in your dataset? 

We checked if it has any correlation with unit_sales, store_nbr, item_nbr or onpromotion. 
We conclude it is retour and left it in the dataset.


```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load CSV file into a DataFrame
df = pd.read_csv('Processed/data_clean.csv')
negative_sales_subset = df[df['unit_sales'] < 0]


# Detected negative sales
print(f"There are {len(negative_sales_subset)} negative sales.")


# Does it have anything to do with unit_sales, store_nbr, item_nbr or onpromotion?
plt.figure(figsize=(10, 6))
plt.subplot(2, 2, 1)
negative_sales_subset['unit_sales'].value_counts().plot(kind='bar', title='Column Value Counts')
plt.subplot(2, 2, 2)
negative_sales_subset['store_nbr'].value_counts().plot(kind='bar', title='Store Number Value Counts')
plt.subplot(2, 2, 3)
negative_sales_subset['item_nbr'].value_counts().plot(kind='bar', title='Item Number Value Counts')
plt.subplot(2, 2, 4)
negative_sales_subset['onpromotion'].value_counts().plot(kind='bar', title='On Promotion Value Counts')
plt.tight_layout()
plt.show()
```
