```python
import pandas as pd

# Clean file

# Read the oil(india) file
with open('Pre-processed/oil(india).csv', 'r') as file:
    lines = file.readlines()

# Replace the comma with a semicolon in the first line
lines[0] = lines[0].replace(',', ';')

# Write the modified lines back to the file
with open('Pre-processed/oil(india).csv', 'w') as file:
    file.writelines(lines)


# Load CSV file into a DataFrame
df_data = pd.read_csv('Pre-processed/data.csv')
df_events = pd.read_csv('Pre-processed/holidays_events(india).csv', delimiter=';')
df_items = pd.read_csv('Pre-processed/items.csv', delimiter=';', encoding='ISO-8859-1')
df_oil = pd.read_csv('Pre-processed/oil(india).csv', delimiter=';', encoding='ISO-8859-1')
df_stores = pd.read_csv('Pre-processed/stores.csv', delimiter=';', encoding='ISO-8859-1')
df_test = pd.read_csv('Pre-processed/test.csv')
df_trans = pd.read_csv('Pre-processed/transactions.csv')


# Print structure and head for each DataFrame
dfs = [df_data, df_events, df_items, df_oil, df_stores, df_test, df_trans]
df_names = ['df_data', 'df_events', 'df_items', 'df_oil', 'df_stores', 'df_test', 'df_trans']

for df, df_name in zip(dfs, df_names):
    print(f"\nDataFrame: {df_name}")
    print("Structure:")
    print(df.dtypes)
    print("\nFirst 10 rows:")
    print(df.head(10))
```
